```python
# Import libraries
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# Load the tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained("sshleifer/distilbart-cnn-12-6")
model = AutoModelForSeq2SeqLM.from_pretrained("sshleifer/distilbart-cnn-12-6")

# Sample article (you can replace this with any long text)
article = """
Artificial intelligence is rapidly changing industries worldwide.
Companies use AI to automate tasks and analyze massive datasets.
In healthcare, AI helps detect diseases early and improve treatment plans.
In finance, it identifies fraudulent transactions efficiently.
However, concerns about job displacement and ethical risks remain.
Experts believe AI will both replace and create jobs in the future.
Governments are working on regulations to ensure responsible AI usage.
"""

# Prepare the input for the model
inputs = tokenizer(
    [article],
    max_length=1024, # BART models typically accept up to 1024 tokens
    truncation=True,
    return_tensors="pt"
)

# Generate summary
summary_ids = model.generate(
    inputs["input_ids"],
    num_beams=4, # Beam search for better quality
    max_length=60,   # maximum summary length
    min_length=25,   # minimum summary length
    early_stopping=True # Stop when all hypotheses have met their max length
)

# Decode and print the summary
summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Print results
print("Original Article:\n")
print(article)

print("\nGenerated Summary:\n")
print(summary_text)
```

**🧩 Explanation (Beginner-Friendly)**

We use a pre-trained summarization pipeline
The model reads the full text and generates a shorter version
max_length → controls how long the summary can be
min_length → prevents very short summaries
do_sample=False → gives consistent results

**BART / T5 (better for summarization)**

```python
# Install first:
# pip install transformers torch

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# Load the tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

# Input article
article = """
Artificial intelligence is transforming industries across the globe.
Companies use AI to automate processes and analyze data.
In healthcare, AI helps detect diseases early.
In finance, it detects fraud efficiently.
However, concerns about job loss and ethics remain.
"""

# Prepare the input for the model
inputs = tokenizer(
    [article],
    max_length=1024, # BART models typically accept up to 1024 tokens
    truncation=True,
    return_tensors="pt"
)

# Generate summary
summary_ids = model.generate(
    inputs["input_ids"],
    num_beams=4, # Beam search for better quality
    max_length=50,   # maximum summary length
    min_length=20,   # minimum summary length
    early_stopping=True # Stop when all hypotheses have met their max length
)

# Decode and print the summary
summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Output
print("BART Summary:\n", summary_text)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

# Load the tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained("t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("t5-small")

# Important: Add prefix "summarize:"
article = """
summarize: Artificial intelligence is transforming industries.
It helps in healthcare and finance but raises ethical concerns.
"""

# Prepare the input for the model
inputs = tokenizer(
    [article],
    max_length=512, # T5 models often use 512 for input length
    truncation=True,
    return_tensors="pt"
)

# Generate summary
# 'do_sample=False' from the original pipeline implies a non-sampling strategy
# For direct model.generate, this typically means using beam search (num_beams > 1)
summary_ids = model.generate(
    inputs["input_ids"],
    num_beams=4, # Use beam search for quality
    max_length=50,
    min_length=20,
    early_stopping=True # Stop when all hypotheses have met their max length
)

# Decode and print the summary
summary_text = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

# Output
print("T5 Summary:\n", summary_text)
```
